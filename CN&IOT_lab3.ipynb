```python
import pandas as pd
```

```python
df=pd.read_csv('as3q2.csv')
```

```python
source=list()
for x in df['Source']:
    if x not in source:
        source.append(x)
```

```python
print(source)
```

```python
dest=list()
for x in df['Destination']:
    if x not in dest:
        dest.append(x)
print(dest)
```

```python
import regex as re
s_port=list()
d_port=list()
for x in df['Info']:
    s_port.append(re.findall(r"(.*) >",x))
```

```python
s_port
```

```python
for x in s_port:
    if len(x)<1:
        s_port.remove(x)
```

```python
print(s_port)
print(len(s_port))
```

```python
for x in df['Info']:
    d_port.append(re.findall(r">(.*)\[",x))
```

```python
d_port
```

```python
for x in d_port:
    if len(x)<1:
        d_port.remove(x)
```

```python
print(d_port)
print(len(d_port))
```
